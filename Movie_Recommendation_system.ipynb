```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

country_input = input("Enter your preferred country: ").lower()
genre_input = input("Enter your preferred movie genre: ").lower()
duration_input = input("Enter your preferred movie duration: ").lower()
release_year_input = input("Enter your preferred release year: ").lower()

df['release_year'] = df['release_year'].astype(str)
df = df.fillna('')

tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df['listed_in'] + ' ' + df['country'] + ' ' + df['type'] + ' ' + df['duration'] + ' ' + df['release_year'])

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

def get_recommendations(title):
    idx = df.index[df['title'] == title].tolist()
    if not idx:
        return []

    idx = idx[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:21]
    movie_indices = [i[0] for i in sim_scores]
    recommendations = df['title'].iloc[movie_indices].tolist()
    recommendations = [rec for rec in recommendations if rec not in filtered_movies['title'].tolist()]
    return recommendations[:10]
filtered_movies = df[
    (df['country'].str.lower().str.contains(country_input)) &
    (df['listed_in'].str.lower().str.contains(genre_input)) &
    (df['duration'].str.lower().str.contains(duration_input)) &
    (df['release_year'].str.lower().str.contains(release_year_input))
]

if filtered_movies.empty:
    filtered_movies = df[df['listed_in'].str.lower().str.contains(genre_input)]
    if not filtered_movies.empty:
        print("No exact matches found. Here are similar movies based on genre:")

for title in filtered_movies['title']:
    recommendations = get_recommendations(title)
    print(f"Recommended movies for {title}:")
    for rec_title in recommendations:
        print(f"- {rec_title}")

if filtered_movies.empty:
    print("No movies found matching your criteria.")
```
